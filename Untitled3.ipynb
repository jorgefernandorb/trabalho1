```python
from ucimlrepo import fetch_ucirepo
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from matplotlib.colors import ListedColormap
# Fetch Iris dataset
iris = fetch_ucirepo(id=53)
X_iris = iris.data.features
y_iris = iris.data.targets.values.ravel() 

# Fetch Vertebral Column dataset
vertebral_column = fetch_ucirepo(id=212)
X_column = vertebral_column.data.features
y_column = vertebral_column.data.targets.values.ravel()  

print('Iris Metadata:', iris.metadata)
print('Iris Variables:', iris.variables)
print('Vertebral Column Metadata:', vertebral_column.metadata)
print('Vertebral Column Variables:', vertebral_column.variables)

# Função para treinar e testar usando KNN
def knn_classifier(X, y, n_neighbors=3):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    return accuracy, conf_matrix

# Função para treinar e testar usando DMC (usando KNN com n_neighbors=1)
def dmc_classifier(X, y):
    return knn_classifier(X, y, n_neighbors=1)

# Função para avaliar o classificador
def evaluate_classifier(classifier_func, X, y, n_runs=20):
    accuracies = []
    for _ in range(n_runs):
        accuracy, _ = classifier_func(X, y)
        accuracies.append(accuracy)
    mean_accuracy = np.mean(accuracies)
    std_accuracy = np.std(accuracies)
    return mean_accuracy, std_accuracy

# Avaliar KNN e DMC nos conjuntos de dados Iris e Coluna
mean_acc_knn_iris, std_acc_knn_iris = evaluate_classifier(knn_classifier, X_iris, y_iris)
mean_acc_dmc_iris, std_acc_dmc_iris = evaluate_classifier(dmc_classifier, X_iris, y_iris)

mean_acc_knn_column, std_acc_knn_column = evaluate_classifier(knn_classifier, X_column, y_column)
mean_acc_dmc_column, std_acc_dmc_column = evaluate_classifier(dmc_classifier, X_column, y_column)

print(f'KNN Iris: Mean Accuracy = {mean_acc_knn_iris}, Std = {std_acc_knn_iris}')
print(f'DMC Iris: Mean Accuracy = {mean_acc_dmc_iris}, Std = {std_acc_dmc_iris}')
print(f'KNN Column: Mean Accuracy = {mean_acc_knn_column}, Std = {std_acc_knn_column}')
print(f'DMC Column: Mean Accuracy = {mean_acc_dmc_column}, Std = {std_acc_dmc_column}')
```

```python


def plot_decision_surface(X, y, classifier, title):
    # Converter rótulos das classes para valores numéricos
    le = LabelEncoder()
    y = le.fit_transform(y)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    classifier.fit(X_train, y_train)
    
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(('red', 'green', 'blue')))
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', edgecolor='k', s=20, label='Training data')
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, marker='x', s=50, label='Test data')  # Removido edgecolor
    plt.title(title)
    plt.legend()
    plt.show()

# Plotar para o conjunto Iris (usando as duas primeiras features)
plot_decision_surface(X_iris.iloc[:, :2].values, y_iris, KNeighborsClassifier(n_neighbors=3), 'KNN Decision Surface for Iris')
plot_decision_surface(X_column.iloc[:, :2].values, y_column, KNeighborsClassifier(n_neighbors=3), 'KNN Decision Surface for Vertebral Column')
```

```python
# Função para calcular e exibir a matriz de confusão
def display_confusion_matrix(X, y, classifier):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    conf_matrix = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:')
    print(conf_matrix)

# Exibir a matriz de confusão para o conjunto de dados da coluna vertebral
display_confusion_matrix(X_column, y_column, KNeighborsClassifier(n_neighbors=3))
```

```python
# Plotar para o conjunto de dados da coluna vertebral (usando as duas primeiras features)
plot_decision_surface(X_column.iloc[:, :2].values, y_column, KNeighborsClassifier(n_neighbors=3), 'KNN Decision Surface for Vertebral Column')
```
